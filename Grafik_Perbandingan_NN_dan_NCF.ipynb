Impor library yang diperlukan

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

Masukkan data dari tabel ke dalam struktur data Python, seperti list atau array numpy

```python
models = ['Neural Network (NN)', 'Neural Collaboration Filtering (NCF)']
mae = [0.794, 0.306]
rmse = [0.960, 0.371]
```

Kita akan membuat grafik batang untuk membandingkan MAE dan RMSE dari kedua model

```python
n_groups = len(models)

# Membuat index untuk setiap grup dan lebar batang
index = np.arange(n_groups)
bar_width = 0.35

fig, ax = plt.subplots()
bar1 = ax.bar(index, mae, bar_width, label='MAE')
bar2 = ax.bar(index + bar_width, rmse, bar_width, label='RMSE')

ax.set_xlabel('Model')
ax.set_ylabel('Scores')
ax.set_title('MAE and RMSE Comparison of Models')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(models)
ax.legend()

plt.show()
```
